```python
import csv

f = open('./data/seoul_book.csv')
data = csv.reader(f)
#data

# for row in data:
#     print(row)

next(data)

high = [] # 최고 기온 담기
low = [] # 최저 기온 담기

for row in data:
    if row[-1] != " "and row[-2]!=" ": # 최고기온(row[-1])과 최저기온(row[-2]) 값이 있다면
        date = row[0].split('-')        # 날짜 값을 '-' 문자 기준으로 구분 --> data 저장
        if 1988 <= int(date[0]):        # 1988년 이후 데이터라면,
            if date[1] =='10' and date[2] == '26': # 10월 26일이라면
                high.append(float(row[-1]))    # 최고 기온값을 high=[] 빈 리스트에 담기
                low.append(float(row[-2]))     # 최저 기온값을 low=[] 빈 리스트에 담기
                
                
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family  = 'Malgun Gothic')
```

```python
plt.rcParams['axes.unicode_minus']=False # 마이너스 기호 깨짐 방지
```

```python
plt.title('PD Revil 생일 기온 변화 그래프 현황')

plt.plot(high, 'red', label="high")
plt.plot(low, 'skyblue', label="low")
plt.legend()
plt.show()
```
