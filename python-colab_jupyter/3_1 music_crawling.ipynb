```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
```

```python
ser = Service('./pandas/chromedriver/chromedriver.exe')
driver = webdriver.Chrome(service = ser)
```

```python
url = 'https://music.bugs.co.kr/chart'
driver.get(url)
```

```python
html = driver.page_source

soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
songs = soup.select('table.byChart > tbody > tr')
len(songs)

song = songs[0]
```

```python
title = song.select('p.title > a')[0].text
title
```

```python
singer = song.select('p.artist > a')[0].text
singer
```

```python
# ['서비스', '순위', '타이틀', '가수']

song_list = []
rank = 1
songs = soup.select('table.byChart > tbody > tr')

for song in songs:
    title = song.select('p.title > a')[0].text
    singer = song.select('p.artist > a')[0].text
    print(title, singer, sep = '|')
    mylist = ['Bugs', rank, title, singer]
    song_list.append(mylist)
    rank += 1

print(song_list)
```

```python
ser = Service('./pandas/chromedriver/chromedriver.exe')
driver = webdriver.Chrome(service = ser)
url = 'https://www.genie.co.kr/chart/top200'
driver.get(url)
```

```python
songs = soup.select('tbody > tr')
# len(songs)
songs[0].select('a.title.ellipsis')[0].text.replace('\n','').strip()
songs[0].select('a.artist.ellipsis')[0].text
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
```

```python
ser = Service('./pandas/chromedriver/chromedriver.exe')
driver = webdriver.Chrome(service = ser)
url = 'https://www.genie.co.kr/chart/top200'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
songs = soup.select('tbody > tr')
len(songs)
song = songs[0]
song
```

```python
title = song.select('a.title')[0].text.strip()
title
```

```python
song_list = []
songs = soup.select('tbody > tr')
rank = 1
for song in songs:
    title = song.select('a.title')[0].text.strip()
    artist = song.select('a.artist')[0].text
    mylist = ['genie', rank, title, artist]
    rank += 1
    song_list.append(mylist)
print(song_list)
```
